```python
import ast
def str_list(x):
  x=ast.literal_eval(x)
  l=[]
  for item in x:

    l.append(item['name'])
  
  return l
```

```python
movies["cast"]=movies["cast"].apply(str_list)
```

```python
#for director(small dataset from movie dataframe)
def fetch_movie_director(obj):
  myArray=[]
  for i in ast.literal_eval(obj):
    if i["job"]=="Director":
      myArray.append(i["name"])
      break
  return myArray
```

```python
movies['crew']=movies["crew"].apply(fetch_movie_director)
```

```python
movies["overview"][1]
```

```python
movies["genres"]=movies['genres'].apply(str_list)
movies["keywords"]=movies['keywords'].apply(str_list)
```

```python
# removing space 
def _join(x):
  l=[]
  for i in x:
    l.append(i.replace(" ",""))
  return l
```

```python
movies["cast"]=movies["cast"].apply(_join)
movies["crew"]=movies["crew"].apply(_join)
```

```python
movies["cast"]=movies["cast"].apply(lambda i:i[0:3])
```

```python
# to convert string into list of strings 
def spliting(text):
  text=str(text).split()
  return text
```

```python
new_df=pd.DataFrame(movies,columns=['id','homepage','title','cast','crew','overview','genres','keywords'])
```

```python
new_df
```

```python
new_df['crew']=new_df['crew'].apply(_join)
```

```python
#makes string to a list of strings
def spliting(text):
  text=str(text).split()
  return text
```

```python
new_df['overview']=new_df['overview'].apply(spliting)
```

```python
#creates tag for every movie
new_df['tags']=new_df['overview'] + new_df['genres'] + new_df['keywords'] + new_df['cast'] + new_df['crew']
```

```python
new_df.drop(columns=['genres' ,'keywords','cast' , 'crew'],inplace=True)
```

```python
def convert4(text):
  l=[]
  for item1 in text:
    l.append(item1)
  return l
```

```python
new_df['tags']=new_df['tags'].apply(convert4)
new_df
```

```python
new_df.head()
```

```python
import nltk
from nltk.stem.porter import PorterStemmer
#converts words of movie dataframe to a sequence to shorten its lookup
ps=PorterStemmer()
def steming(text):
  l=[]
  for i in text:

    l.append(ps.stem(i))
  return l
```

```python
new_df['tags']=new_df['tags'].apply(steming)
```

```python
new_df["tags"]=new_df["tags"].apply(_join)
```

```python
new_df['tags'] = new_df['tags'].apply(lambda x: " ".join(x))
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(stop_words='english')
```

```python
#converts list of strs to array
data = cv.fit_transform(new_df['tags']).toarray()
```

```python
data
```

```python
#import cosine similarity to see the similarity between new_df and movie dataframe
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity = cosine_similarity(data)
```

```python
similarity
```

```python
similarity.shape
```

```python
#converts data object to enumerate object 
sorted(list(enumerate(similarity[0])),reverse=True,key = lambda x: x[1])[1:6]
```

```python
def make_recommendation(movie):
  #checks if new_df and movie dataframe has same title
    index = new_df[new_df['title'] == movie].index[0]
    #distance between two vectors(new_df,movies)
    distances = sorted(list(enumerate(similarity[index])),reverse=True,key = lambda x: x[1])
    for i in distances[1:8]:

      #print movies id and new_df for specific distance and index
      print("{} {}".format(new_df.iloc[i[0]].title,new_df.iloc[i[0]]))
```

```python
make_recommendation("Avatar")
```

```python
make_recommendation("Batman")
```
